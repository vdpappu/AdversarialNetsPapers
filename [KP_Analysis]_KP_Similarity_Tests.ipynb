```
import os
os.chdir("/content/drive/My Drive/PhiGRC_Dumps/key_phrase_experiments")

import pandas as pd
import pickle
import numpy as np
from scipy.spatial.distance import cosine
```

```
sample_kp_feats = pickle.load(open("key_phrase_feats.pkl","rb"))
kp_graph = pickle.load(open("key_phrase_graph.pkl","rb"))
kp_graph = kp_graph.subgraph(list(sample_kp_feats))
```

```
edge_freq_list = [len(kp_graph[ele]) for ele in list(kp_graph)]
edge_freq_df = pd.DataFrame({'key_phrase': list(kp_graph),
                             "edge_freq": edge_freq_list}).sort_values(by="edge_freq", ascending=False).reset_index(drop=True)
```

**Change the candidate_phrase in below cell to desired phrase**

```
#utility to get top-n phrases to a given phrase
#fetch all one hop phrases and search
candidate_phrase = "availability of credit" #input search key-phrase here
edge_list = list(kp_graph[candidate_phrase])
search_space = []
for edge in edge_list:
  search_space+=list(kp_graph[edge])
search_space+=edge_list
search_space = list(set(search_space))

dist_list = []
for kp in list(sample_kp_feats):
  dist_list.append(1-cosine(sample_kp_feats[candidate_phrase],sample_kp_feats[kp]))

dist_df = pd.DataFrame({'key_phrase': list(sample_kp_feats),
                        'distance': dist_list}).sort_values(by='distance', ascending=False).reset_index(drop=True)
dist_df['num_toks'] = dist_df['key_phrase'].apply(lambda x: len(x.split(' ')))
dist_df = dist_df[dist_df['num_toks']>1]
del dist_df['num_toks']
```
